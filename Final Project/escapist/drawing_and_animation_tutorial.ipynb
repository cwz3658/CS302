```python
"""
Patches
   1.Primative shapes in Matplotlib are known as patches
    2.Provided by the patches module
    3.Subclasses of patch provide implementations for Rectangles, Arrow, Ellipses, Arcs, Circles
    4.All of this is part of the Artist API, which also provides support for text.
    5.In fact, everything drawn using Matplotlib is part of the artists module.
    6.It's just a different level of access for drawing shapes compared to plots. 




Drawing
    1.Drawing is a matter of adding the patch to the current figure's axes

"""
```

```python
import matplotlib.pyplot as plt
plt.axes()

circle = plt.Circle((0, 0), radius = 0.75, fc='y')
plt.gca().add_patch(circle)  #gca() returns the current Axis instance.

plt.axis('scaled')
plt.show()
```

```python
# rectangles
plt.axes()
rectangle = plt.Rectangle( (10, 10), 100, 100, fc='r' )

plt.gca().add_patch(rectangle)  #gca() returns the current Axis instance.

plt.axis('scaled')
plt.show()
```

```python
# rectangles
plt.axes()
circle = plt.Circle( (0, 0), 0.75, fc='y' )

plt.gca().add_patch(circle)  #gca() returns the current Axis instance.

plt.axis('scaled')
plt.show()
```

```python
# Lines
plt.axes()
dotted_line = plt.Line2D((2, 8), (4, 4), lw=5., 
                         ls='-.', marker='.', 
                         markersize=50, 
                         markerfacecolor='r', 
                         markeredgecolor='r', 
                         alpha=0.5)
plt.gca().add_line(dotted_line)

plt.axis('scaled')
plt.show()
```

```python
# Animation:
#   Move certain shapes around
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

fig = plt.figure()     # create a new figure object, which is the window showing up

# set some properties
fig.set_dpi(100) 
fig.set_size_inches(7, 6.5)

ax = plt.axes(xlim=(0, 10), ylim=(0, 10))
patch = plt.Circle((5, -5), 0.75, fc='y')

def init():
    patch.center = (5, 5)
    ax.add_patch(patch)
    return patch,

def animate(i):
    x, y = patch.center
    x = 5 + 3 * np.sin(np.radians(i))
    y = 5 + 3 * np.cos(np.radians(i))
    patch.center = (x, y)
    return patch,

anim = animation.FuncAnimation(fig, animate, 
                               init_func=init, 
                               frames=360, 
                               interval=20,
                               blit=True)

plt.show()
```

```python

# save animation
anim.save('animation.mp4', fps=30, 
          extra_args=['-vcodec', 'h264', 
                      '-pix_fmt', 'yuv420p'])
```
